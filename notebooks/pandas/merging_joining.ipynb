## What is iloc?

iloc is used for integer position-based indexing.

It selects data based on:
- Row position (0,1,2…)
- Column position (0,1,2…)

Syntax:
df.iloc[row_index, column_index]


## Difference Between loc and iloc

| Feature | loc | iloc |
|--------|-----|------|
| Basis | Labels | Positions |
| Input | Index names | Integer index |
| Usage | Named selection | Numeric selection |


```python
# Select a single row using iloc
row_iloc = student_data.iloc[0]
row_iloc
```

```python
# Select specific columns using iloc
row_iloc = df.iloc[0:9]
row_iloc
```

```python
# Slice rows and columns using iloc
sliced = df.iloc[:5, 0:3]
sliced
```

# Merging, Joining, and Concatenating DataFrames in Pandas

In real-world data analysis, data is rarely stored in a single file.
Instead, data is often distributed across:
- Multiple CSV files
- Different Excel sheets
- Various databases
- APIs and logs

To perform proper analysis, we must combine multiple datasets.
This process is called:
- Merging
- Joining
- Concatenation


## Why Combining DataFrames is Important?

In real-life scenarios:
- Sales data may be in one file
- Customer data in another file
- Product data in a separate file

To analyze complete information, we need to combine these datasets.

This is widely used in:
- Business Analytics
- Machine Learning pipelines
- Data Engineering workflows
- Data Cleaning & EDA


## Three Main Ways to Combine Data in Pandas

1. Concatenation (concat) → Stack DataFrames vertically or horizontally  
2. Merging (merge) → Combine based on common keys (like SQL JOIN)  
3. Joining (join) → Combine DataFrames based on index  

Each method serves a different purpose.


## What is Concatenation?

Concatenation means:
- Stacking DataFrames together
- Along rows (vertical)
- Along columns (horizontal)

It is useful when:
- Data is split across multiple files
- Monthly datasets need to be combined


```python
# Import pandas library


# Import numpy library
```

We will create multiple DataFrames to simulate real-world datasets

such as sales data from different months.


```python
# Create first DataFrame (e.g., Month 1 data)
df_1 = pd.DataFrame({
    "Id": [1,2,3],
    "Name": ["Maruti", "hyundai", "Honda"],
    "Price": [300000,400000,500000]
})

# Create second DataFrame (e.g., Month 2 data)
df_2 = pd.DataFrame({
    "Id": [4,5,6],
    "Name": ["Suzuki", "Toyota", "Ford"],
    "Price": [400000,300000,600000]
})

df_1, df_2
```

Vertical concatenation stacks DataFrames on top of each other.

Used when columns are the same but rows are different.


```python
# Concatenate DataFrames vertically
vertical_concat = pd.concat([df_1,df_2], axis = 0)
vertical_concat
```

Horizontal concatenation combines DataFrames side by side.

Used when rows represent the same entities but columns differ.


```python
# Concatenate DataFrames horizontally
df_f = pd.DataFrame({
    "Mileage":[20,16,17],
    "Fuel":["Petrol", "Diesel","Petrol"]
})

horizonatal_concat = pd.concat(
    [df_1, df_f], axis = 1
    )
horizonatal_concat
```

## What is Merging?

Merging is similar to SQL JOIN operations.

It combines two DataFrames based on a common column (key).
Example:
- Student Data (ID, Name)
- Marks Data (ID, Score)

We merge them using the common key: ID


## Types of Merge Operations in Pandas

1. Inner Merge → Only matching records
2. Left Merge → All records from left + matches from right
3. Right Merge → All records from right + matches from left
4. Outer Merge → All records from both DataFrames


```python
df_1 = pd.DataFrame({
    "Id": [1,2,3],
    "Name": ["Maruti", "hyundai", "Honda"],
    "Price": [300000,400000,500000]
})

# Create second DataFrame (e.g., Month 2 data)
df_2 = pd.DataFrame({
    "Id": [1,2,5],
    "Mileage":[20,16,17],
    "Fuel":["Petrol", "Diesel","Petrol"]
    
})
```

```python
df_1
```

```python
df_2
```

Inner merge returns only the rows where keys exist in both DataFrames.


```python
# Merge two DataFrames using inner join on common key
inner_merge = pd.merge(df_1,df_2, on = "Id", how = "inner")
inner_merge
```

Left merge keeps all rows from the left DataFrame

and matches rows from the right DataFrame when available.


```python
# Perform left merge using a common key
left_merge = pd.merge(df_1, df_2, on = "Id", how = "left")
left_merge
```

Right merge keeps all rows from the right DataFrame

and matches rows from the left DataFrame when available.


```python
# Perform right merge using a common key
right_merge = pd.merge(df_1,df_2, on = "Id", how = "right")
right_merge
```

Outer merge returns all rows from both DataFrames.

Missing matches are filled with NaN values.


```python
# Perform outer merge on the common key
outer_merge = pd.merge(df_1,df_2, on = "Id", how = "outer")
outer_merge
```

## What is a Pivot Table?

A Pivot Table is used to summarize and aggregate data dynamically.

It helps to:
- Analyze relationships between variables
- Generate summary statistics
- Create analytical reports

Similar to Excel Pivot Tables.


## Why Pivot Tables are Important?

Pivot tables allow:
- Multi-dimensional aggregation
- Quick business insights
- Flexible data analysis

Widely used in:
- Sales analysis
- Financial reporting
- Market analytics
- EDA workflows


```python
df
```

```python
# Create a pivot table to summarize a numerical column by a category
pivot_fuel = pd.pivot_table(
    df,
    values = "selling_price",
    index ="fuel",
    aggfunc= "mean"
    
)
pivot_fuel
```

```python
# Create a pivot table to summarize a numerical column by a category
pivot_fuel = pd.pivot_table(
    df,
    values = "selling_price",
    index ="fuel",
    aggfunc= ['mean','max','count']
    
)
pivot_fuel
```

```python
# Create pivot table using multiple index and columns
pivot_fuel_transmission = pd.pivot_table(
    df,
    values ="selling_price",
    index = "fuel",
    columns = "transmission",
    aggfunc= "mean"
)
pivot_fuel_transmission
```

## What is Crosstab?

Crosstab (Cross Tabulation) is used to:
- Calculate frequency counts
- Analyze relationships between categorical variables
- Create contingency tables

It is commonly used in:
- Survey analysis
- Market research
- Categorical data analysis


## Crosstab vs Pivot Table

| Feature | Pivot Table | Crosstab |
|--------|-------------|----------|
| Purpose | Aggregation | Frequency counts |
| Data Type | Numeric + Categorical | Mostly categorical |
| Flexibility | High | Simpler frequency analysis |


```python
# Create a crosstab between two categorical columns
crosstab = pd.crosstab(df['fuel'], df['transmission'])
crosstab
```

```python
seller = pd.crosstab(
    df['seller_type'],
    df['owner']
)
seller
```

```python
# Create crosstab with normalization or aggregation
price = pd.crosstab(
    df['fuel'],
    df["transmission"],
    values = df['selling_price'],
    aggfunc="mean"
)
price
```
