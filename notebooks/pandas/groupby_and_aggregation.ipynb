## What is Grouping in Pandas?

Grouping means:
- Splitting data into groups based on some category
- Applying calculations on each group
- Combining results into summarized insights

In Pandas, grouping is done using:
groupby()

This follows the concept:
Split → Apply → Combine


## Real-Life Examples of Grouping

Grouping is widely used in:
- Business Analytics (sales by region)
- Education (marks by subject)
- Finance (expenses by category)
- Automotive Analysis (price by car brand)

For example:
Instead of analyzing 200 cars individually,
we can analyze:
Average price per car manufacturer.


Pandas groupby() is an essential method for data aggregation and analysis in python. It follows the "Split-Apply-Combine" pattern, which means it allows users to −

- Split data into groups based on specific criteria.
- Apply functions independently to each group.
- Combine the results into a structured format.

Every groupby() operation involves three key steps, splitting data into groups based on some criteria, apply functions independently to each group, and then merge the results back into a meaningful structure.

In many situations, we apply some functions on each splitted groups. In the apply functionality, we can perform the following operations −

- Aggregation: Computing summary statistics like mean, sum, etc.
- Transformation: Applying a function to transform data.
- Filtration: Removing groups based on some condition.

## Understanding groupby() Function

The groupby() function is used to:
- Group rows based on a categorical column
- Perform aggregation operations on each group

Syntax:
df.groupby("column_name")


```python
# Group the dataset by a categorical column (e.g., fuel)
fuel_group = df_car.groupby("fuel")
fuel_group
```

```python
fuel_group.head()
```

## Aggregation in Pandas

Aggregation means applying statistical functions such as:
- sum()
- mean()
- median()
- min()
- max()
- count()

These help summarize large datasets efficiently.


```python
# Calculate mean of numerical columns after grouping
mean_by_fuel = fuel_group[['selling_price', 'km_driven']].mean()
mean_by_fuel

```

```python
# Apply multiple aggregation functions (mean, max, min)
agg_multiple = fuel_group['selling_price'].agg(['mean', 'max', 'min', 'count'])
agg_multiple
```

## Grouping by Multiple Columns

We can group data using more than one categorical column.

Example:
Group by:
- fuel
- owner

This creates hierarchical grouping.


```python
# Group dataset by multiple columns
pd.set_option('display.float_format', '{:.2f}'.format)

multi_group = df_car.groupby(['fuel', 'owner'])

multi_group_mean = multi_group['selling_price'].mean()

multi_group_mean
```

## Selecting Specific Columns After Grouping

Instead of aggregating the whole dataset,
we can select specific numerical columns for analysis.
This improves clarity and performance.


```python
# Group by a category and aggregate only selected column
fuel_group = df_car.groupby("fuel")[['selling_price', 'year', 'km_driven']].mean()
fuel_group

```

## Group Size and Counting Records

Sometimes we want to know:
- Number of records per group
- Frequency of categories

This helps in understanding data distribution.


```python
# Count number of records in each group
group_counts = df_car.groupby("transmission").size()
group_counts
```

## Using agg() Function (Advanced Aggregation)

The agg() function allows:
- Applying multiple operations at once
- Custom aggregation logic
- Better summarized reports


```python
# Use agg() to calculate multiple statistics per group
adv_agg = df_car.groupby("fuel").agg(
    {
        "selling_price": ['mean', 'max','min'],
        "km_driven": ['mean', 'max'],
        "year": ["mean"]
    }
)


adv_agg
```

## Transform vs Aggregate (Important Concept)

Aggregation:
- Reduces data size
- Produces summary values

Transform:
- Keeps original data size
- Applies group-wise calculations

This is very important in feature engineering.


```python
df_car.head()
```

```python
# Apply group-wise transformation on a numerical column

df_car['avg_sell_price'] = df_car.groupby("fuel")['selling_price'].transform("mean")

#display relevant columns 
df_car[['fuel', 'selling_price', 'avg_sell_price']].head()
```

```python
df_car.head()
```

## Filtering Groups

We can filter entire groups based on a condition.
Example:
- Only show groups with average price > threshold


```python
# Filter grouped data based on a condition
filter_group = df_car.groupby("fuel").filter(
    lambda x:x['selling_price'] .mean()> 500000
)
filter_group
```

## Sorting Grouped Results

After aggregation, results can be sorted to:
- Identify top categories
- Find highest or lowest performing groups


```python
# Sort grouped aggregation results by a numerical metric
sorted_df = df_car.groupby("owner")['selling_price'].mean().sort_values(ascending= False)
sorted_df
```

## Common Mistakes in Grouping (Important)

1. Grouping without cleaning missing values  
2. Aggregating non-numeric columns  
3. Forgetting reset_index() after groupby  
4. Misinterpreting grouped results  

Always inspect grouped data carefully.


1. Which fuel type has highest average resale price?
2. Do automatic cars sell for more than manual cars?
3. Which year group (older vs newer) has higher average price?
4. Which seller type dominates the dataset?
