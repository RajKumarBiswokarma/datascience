## Applying Functions

Pandas allows applying custom functions using:
- apply()
- map()
- lambda functions

This is useful for feature transformation.


```python
# Apply a custom function to transform the price column
def price_in_thousands(x):
    if pd.notnull(x):
        return x / 1000
    return x 

df["price_in_thousands"] = df['price'].apply(price_in_thousands)

df.head()

```

```python
# Use lambda function to modify a numeric column (e.g., scaling price)
df['horsepower_scaled'] = df['horsepower'].apply(lambda x: x * 1.1 if pd.notnull(x) else x)

df[['horsepower', 'horsepower_scaled']].head()
```

## Creating New Columns

We can create new columns based on existing data.
This is commonly used in data preprocessing.


```python
# Create a new column such as price_category based on price values
def price_category(price):
    if pd.isnull(price):
        return "Unknown"
    elif price <  10000:
        return "Low"
    elif price < 20000:
        return "Medium"
    else: 
        return "High"
    
    
df['price_category'] = df['price'].apply(price_category)

df[['price', 'price_category']].head()
```

```python
df['fuel_type'].unique()
```

```python
# Convert categorical values (e.g., fuel type) into numerical form using mapping
fuel_map = {'gas': 0, 'diesel': 1}

df['fuel_encode'] = df['fuel_type'].map(fuel_map)

df[['fuel_type', 'fuel_encode']].head()
```
