# What is Plotly?

Plotly is a modern Python visualization library used for creating interactive and dynamic visualizations.

Unlike Matplotlib and Seaborn, Plotly graphs are:
- Interactive
- Zoomable
- Hover-enabled
- Presentation-friendly
- Web-compatible

This makes Plotly highly useful in real-world dashboards, business analytics, and data storytelling.

### Why Do We Use Plotly ?

### Interactivity
Plotly allows users to:
- Hover over data points
- Zoom into specific regions
- Toggle graph elements
- Explore data visually

### Better for Presentations & Dashboards
Plotly visualizations are widely used in:
- Business dashboards
- Data reports
- Web applications
- Analytical presentations

### Real-World Relevance
In industry, Plotly is commonly used alongside tools like Dash and Streamlit to build interactive data apps.

## Plotly vs Matplotlib vs Seaborn
Matplotlib:
- Static plots
- Highly customizable
- Used for foundational visualization

Seaborn:
- Statistical plots
- Better aesthetics
- EDA focused

Plotly:
- Interactive visualizations
- Modern UI and tooltips
- Best for dashboards and presentations

# Scatter Plot 

A Scatter Plot is used to visualize the relationship between two numerical variables.
With Plotly, the same plot becomes interactive with hover information and zoom functionality.

```python
# !pip install nbformat 
```

```python
# Create an interactive scatter plot using plotly express
# Select x-axis and y-axis numerical features
# Add color grouping (optional, e.g., species)
# Display the interactive plot

fig = px.scatter(x=[0,1,2,3,4], y = [0,1,4,9,16])
fig.show()
```

```python
fig = px.scatter(iris, x="sepal_width", y= "sepal_length", color='species', title ="Scatter Plot")
fig.show()
```

```python
fig = px.scatter(iris, x="sepal_width", y= "sepal_length", color='species', title ="Scatter Plot", size="petal_length")
fig.show()
```

# Line Plot

A Line Plot is used to observe trends in numerical data.
In Plotly, line plots become interactive, allowing dynamic exploration of trends.

```python
iris.head()
```

```python
# Create an interactive line plot
# Select feature for x-axis (index or category)
# Select numerical feature for y-axis
# Add title and labels
# Display the interactive visualization

fig = px.line(
    data_frame= iris,
    y="sepal_length",
    title = "Line Plot of Sepal Length"
)
fig.show()
```

```python
df = px.data.gapminder().query("country == 'Nepal'")
df
```

```python
fig = px.line(
    data_frame=df,
    x="year",
    y="lifeExp",
    title ="Life Expectancy in Nepal"
)
fig.show()
```

```python
df_continent = px.data.gapminder().query("continent == 'Oceania'")
df_continent
```

```python
fig = px.line(
    data_frame=df_continent,
    x ="year",
    y ="lifeExp",
    color = "country"
)
fig.show()
```

```python
df = px.data.gapminder().query("country in ['Canada', 'Botswana']")
```

```python
fig = px.line(
    data_frame= df,
    x ="lifeExp",
    y="gdpPercap",
    color = "country",
    text="year"
)
fig.update_traces(textposition = "bottom right")
fig.show()
```

#  Bar Plot

Bar plots are used to compare values across different categories.
Plotly bar charts allow hover tooltips showing exact values for better interpretation.

```python
# Create an interactive bar plot
# Select categorical feature for x-axis
# Select numerical feature for y-axis (aggregation concept)
# Add chart title for clarity
# Show the interactive bar chart

fig = px.bar(
    data_frame=iris,
    x ="species",
    y="sepal_length",
    color="species",
    title ="Bar Plot"
)
fig.show()
```

```python
df_nepal = px.data.gapminder().query("country == 'Nepal'")
```

```python
fig =px.bar(df_nepal, x ='year', y = "pop")
fig.show()
```

```python
medal_df = px.data.medals_long()
medal_df
```

```python
fig = px.bar(medal_df, x ="nation", y="count", color="medal", title="Medal bar graph")
fig.show()
```

# Histogram (Interactive Distribution Plot)

Histogram shows the distribution of a numerical variable.
In Plotly, histograms allow dynamic bin visualization and hover insights.

```python
# Create an interactive histogram for a numerical feature
# Select dataset and numerical column
# Customize bins (optional)
# Display the histogram

iris.head()
```

```python
fig = px.histogram(
    iris,
    x="sepal_length",
    title="Histogram ",
    nbins = 20,
    color ="species"
)
fig.show()
```

## Box Plot in Pyplot

```python
# Create an interactive boxplot
df_tips =px.data.tips()
df_tips.head()
```

```python
fig = px.box(
    df_tips,
    y="total_bill",
    title ="BOX Plot of Total Bill"
    
)
fig.show()
```

```python
fig = px.box(
    df_tips,
    x ="time",
    y="total_bill",
    title ="BOX Plot of Total Bill"
    
)
fig.show()
```

```python
fig = px.box(
    df_tips,
    x ="time",
    y="total_bill",
    title ="BOX Plot of Total Bill",
    points ='all'
    
)
fig.show()
```

## Violin Plot in Pyplot 


```python
# Create an interactive Violin plot
fig = px.violin(
    df_tips,
    y="total_bill",
    title ="Violin Plot of Total Bill"
)
fig.show()
```

```python
# Create an interactive Violin plot
fig = px.violin(
    df_tips,
    box = True,
    y="total_bill",
    title ="Violin Plot of Total Bill",
    points = 'all'
)
fig.show()
```
