# 1. Introduction to Pairplot

## What is a Pairplot?

A Pairplot is a multivariate visualization tool in Seaborn that shows pairwise relationships between all numerical features in a dataset.

It automatically creates:
- Scatter plots between feature pairs
- Distribution plots on the diagonal
- A full relationship matrix of the dataset

## Why Pairplot is Important in Data Science

Pairplots help in:
- Detecting correlations between features
- Identifying clusters in data
- Understanding feature relationships
- Observing class separation (when hue is used)
- Detecting outliers visually

It is one of the most commonly used plots during Exploratory Data Analysis (EDA).

## How Pairplot Works 

If a dataset has 4 numerical features, pairplot creates:
- 6 pairwise scatter plots
- 4 diagonal distribution plots
- A grid showing all feature relationships

This gives a complete statistical overview of the dataset in a single visualization.

```python
# Import required libraries for data analysis and visualization
# (NumPy, Pandas, Matplotlib, Seaborn)
import seaborn as sns
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Load the dataset for multivariate visualization (e.g., iris dataset)
iris = sns.load_dataset("iris")


# Preview the dataset structure before visualization
iris.head()
```

```python
# Create a basic pairplot for all numerical features
# Observe scatter plots in off-diagonal positions
# Observe distribution plots on the diagonal

sns.pairplot(iris)
plt.show()
```

```python
sns.pairplot(iris, hue='species')
plt.show()
```

Using the hue parameter allows us to color data points based on a categorical feature such as class labels or species.

This helps in:
- Visualizing class separation

- Feature discrimination analysis

```python
sns.pairplot(iris, hue='species')
plt.show()
```

# 2. Introduction to Boxplot

## What is a Boxplot?

A Boxplot is a statistical visualization that summarizes the distribution of a dataset using five key statistical measures:
- Minimum
- First Quartile (Q1)
- Median (Q2)
- Third Quartile (Q3)
- Maximum

It is also very effective for detecting outliers.

## Why Boxplot is Used in EDA

Boxplots help us:
- Understand data spread
- Identify skewness
- Detect outliers
- Compare distributions across categories
- Analyze feature variability

This makes it extremely useful in feature analysis before model training.

## Components of a Boxplot 

- Box → Interquartile Range (Q1 to Q3)
- Middle Line → Median
- Whiskers → Data range (excluding outliers)
- Points outside whiskers → Outliers

```python
# Create figure and axes using plt.subplots()
fig, ax = plt.subplots(figsize=(8,8))

# Select categorical feature for x-axis (e.g., species)
# Select numerical feature for y-axis (e.g., petal_length)
# Create a seaborn boxplot for statistical distribution analysis
sns.boxplot(data = iris, x ="species", y = "petal_length", ax = ax)

# Add title and axis labels for interpretation4
ax.set_title("Box Plot of Petal Length Acroos Species ")
ax.set_xlabel("Species")
ax.set_ylabel("Petal Length")

# Display the plot
plt.show()
```

## Boxplot for Multiple Feature Comparison

Boxplots can also be used to compare distributions of multiple numerical features across categories.
This helps in identifying which features vary the most across different classes.

```python
# Create figure and axes using plt.subplots()
fig, ax = plt.subplots(figsize=(8,8))

# Select categorical feature for x-axis (e.g., species)
# Select numerical feature for y-axis (e.g., petal_length)
# Create a seaborn boxplot for statistical distribution analysis
sns.boxplot(data = iris, x ="species", y = "sepal_length", ax = ax)

# Add title and axis labels for interpretation4
ax.set_title("Box Plot of sepal_length Acroos Species ")
ax.set_xlabel("Species")
ax.set_ylabel("sepal_length")

# Display the plot
plt.show()
```

# 3. Introduction to Violin Plot

## What is a Violin Plot?

A Violin Plot is an advanced statistical plot that combines:
- Boxplot (statistical summary)
- KDE plot (distribution density)

It provides both distribution shape and statistical spread in a single visualization.

## Why Violin Plot is Powerful

Violin plots help us:
- See full data distribution
- Understand density of values
- Compare distributions across categories
- Detect multimodal distributions

It gives more detailed insight than a boxplot alone.

## Boxplot vs Violin Plot

- Boxplot → Shows summary statistics (median, quartiles, outliers)
- Violin Plot → Shows full distribution + density shape

In real-world EDA:
Boxplot = Quick summary  
Violin Plot = Detailed distribution analysis

```python
# Create figure and axes using plt.subplots()
fig, ax = plt.subplots(figsize=(9,9))

# Select categorical feature for x-axis (e.g., species)
# Select numerical feature for y-axis (e.g., sepal_length)
# Create a violin plot to visualize distribution and density
sns.violinplot(data = iris, x = "species", y ="sepal_length", ax = ax )
# Add title and axis labels

# Display the visualization
ax.set_title("Violin Plot of Sepal length Acrross Species")
ax.set_xlabel("Species")
ax.set_ylabel("Sepal Length")
plt.show()
```

# 4. When to Use Each Plot 

Use Pairplot when:
- multivariate relationship analysis
- feature correlation overview
- performing initial EDA

Use Boxplot when:
- statistical summary of distribution
- outlier detection
- comparing categories

Use Violin Plot when:
- full distribution insight
- density + spread together
- deeper statistical analysis

# 5. Introduction to Correlation Matrix

## What is Correlation?

Correlation is a statistical measure that shows the relationship between two numerical variables.

It tells us:
- How strongly two variables are related
- Whether the relationship is positive or negative
- The strength of linear dependency between features

## Correlation Value Range 

Correlation values always lie between -1 and +1.

- +1 → Perfect positive correlation
-  0 → No correlation
- -1 → Perfect negative correlation

Interpretation:
- Higher positive value → Features increase together
- Higher negative value → One increases while the other decreases
- Near zero → Weak or no linear relationship

## What is a Correlation Matrix?

A Correlation Matrix is a table that shows correlation coefficients between multiple numerical features in a dataset.

Instead of comparing two features manually, it computes correlation for:
- Feature vs Feature (all combinations)
- Entire dataset at once

This makes it a powerful tool for feature analysis in Data Science.

## Important Note: Numerical Features Only

Correlation is calculated only on numerical columns because:
- Mathematical correlation requires numeric values
- Categorical variables must be excluded or encoded

Most libraries automatically ignore non-numeric columns when computing correlation.

```python
iris.head()
```

```python
iris.info()
```

```python
# Select numerical features from the dataset
iris_numeric = iris.select_dtypes(include=['float64', 'int64'])

iris_numeric.head()
```

```python
# Calculate the correlation matrix using dataframe correlation function
corr_matrix = iris_numeric.corr()

# Store the correlation matrix into a new variable
corr_matrix
```

# 6. Introduction to Heatmap

## What is a Heatmap?

A Heatmap is a graphical representation of data where values are represented using colors.

In the context of correlation:
- Each cell represents correlation between two features
- Color intensity represents strength of correlation
- Helps quickly identify strong and weak relationships

## Why Heatmaps are Used in Data Science

Heatmaps help in:
- Identifying highly correlated features
- Detecting multicollinearity
- Feature selection before model training
- Understanding dataset structure visually
- Faster interpretation compared to raw correlation tables

## Heatmap for Correlation Matrix (Core Visualization)

This is one of the most important visualizations in Exploratory Data Analysis (EDA).  
It converts the numerical correlation matrix into a color-coded visual map for easy interpretation.

```python
# Create figure and axes using plt.subplots()
fig, ax = plt.subplots(figsize = (8,8))
# Generate the correlation matrix from the numerical dataset
# corr_matrix

# Create a seaborn heatmap using the correlation matrix
sns.heatmap(corr_matrix, ax = ax)

# Add title to the heatmap for clarity
ax.set_title("Basic correlation heatmap")

# Display the final visualization
plt.show()
```

#  Annotated Heatmap (Detailed Interpretation)

## What is Annotation in Heatmap?

Annotation means displaying the numerical correlation values inside each heatmap cell.

This allows us to:
- See exact correlation values
- Interpret strength of relationships precisely
- Combine visual and numerical analysis together

```python
# Create figure and axes using plt.subplots()
fig, ax = plt.subplots(figsize = (8,8))
# Generate the correlation matrix from the numerical dataset
# corr_matrix

# Create a seaborn heatmap using the correlation matrix
sns.heatmap(corr_matrix, annot = True, ax = ax)

# Add title to the heatmap for clarity
ax.set_title("Annoted  correlation heatmap")

# Display the final visualization
plt.show()
```

# Understanding Color Maps (cmap)

Color maps control how correlation strength is represented visually.

Common Colormaps:
- coolwarm → Best for correlation visualization
- viridis → Smooth gradient visualization
- Blues → Single color intensity scale

Interpretation:
- Darker colors → Strong correlation
- Lighter colors → Weak correlation
- Opposite colors → Negative correlation

```python
# Create figure and axes
fig, ax = plt.subplots(figsize=(8, 6))

# Heatmap with colormap and annotations
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', ax=ax)

# Set title
ax.set_title("Correlation Heatmap with Coolwarm Colormap")

# Display plot
plt.show()
```

#  How to Interpret a Correlation Heatmap 

While analyzing a heatmap, observe:

- Values close to +1 → Strong positive relationship
- Values close to -1 → Strong negative relationship
- Values near 0 → Weak or no relationship

Key EDA Insights:
- Highly correlated features may cause multicollinearity
- Strong relationships help in feature selection
- Weak correlations may indicate independent features


Correlation heatmaps are used in professional ML workflows to:
- Detect multicollinearity
- Perform feature selection
- Remove redundant features
- Improve model performance
- Enhance interpretability during EDA

This is a standard step before model training in real-world Data Science pipelines.
