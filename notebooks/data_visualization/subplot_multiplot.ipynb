## Subplots

In the context of data visualization, subplots refer to a layout within a single figure where multiple plots (or axes) are arranged in rows and columns. 

It is a common and useful task, especially when you want to display multiple plots within the same figure for analysing different aspects of data.



## Why Use Subplots?

Subplots are useful for:
- Comparative analysis
- Dashboard-style visualization
- Multi-feature EDA

## The subplot() Function

The subplot() function takes three arguments that describes the layout of the figure.

The layout is organized in rows and columns, which are represented by the first and second argument.

The third argument represents the index of the current plot.


## Using plt.subplot() Function

plt.subplot(rows, columns, index)

This allows us to arrange multiple plots in a grid layout.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create a 2x2 subplot layout
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

plt.subplot(1,2,1)
plt.plot(x,y)  # plot 1

x = np.array([0,1,2,3])
y = np.array([10,20,30,40])

plt.subplot(1,2,2)
plt.plot(x,y)




# Plot different charts in each subplot
plt.show()
```

```python
# Create a 2x2 subplot layout
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

plt.subplot(1,2,1)
plt.title("First Data")
plt.plot(x,y)  # plot 1

x = np.array([0,1,2,3])
y = np.array([10,20,30,40])

plt.subplot(1,2,2)
plt.title("Second Data")
plt.plot(x,y)

plt.suptitle("SUPER TITLE")



# Plot different charts in each subplot
plt.show()
```

```python
# 2 rows and 3 columns ---> 6 fig

x = np.array([0,1,2,3])
y = np.array([10,20,30,40])

plt.subplot(2,3,1)
plt.plot(x,y)

plt.subplot(2,3,2)
plt.plot(x,y)

plt.subplot(2,3,3)
plt.plot(x,y)

plt.subplot(2,3,4)
plt.plot(x,y)

plt.subplot(2,3,5)
plt.plot(x,y)

plt.subplot(2,3,6)
plt.plot(x,y)


plt.show()
```

## Using plt.subplots() (Advanced and Recommended)

plt.subplots() provides better control over multiple axes.
It is widely used in professional data analysis.

The subplots() function is a part of the Matplotlib library and is available in both matplotlib interfaces, the object-oriented approach (matplotlib.figure.Figure.subplots()) and the functional interface (matplotlib.pyplot.subplots()).

The matplotlib.figure.Figure.subplots() method is called on a Figure object and returns an Axes or array of Axes.

The matplotlib.pyplot.subplots() function is a wrapper for matplotlib.figure.Figure.subplots() and directly creates a figure object along with an axes or array of axes object.

`matplotlib.pyplot.subplots(nrows=1, ncols=1, *, sharex=False, sharey=False, squeeze=True, width_ratios=None, height_ratios=None, subplot_kw=None, gridspec_kw=None, **fig_kw)`

- nrows, ncols: Number of rows and columns of the subplot grid.
- sharex, sharey: Controls sharing of properties among x or y axes.
- squeeze: Controls the squeezing of extra dimensions from the returned array of axes.
- subplot_kw: Dictionary with keywords passed to the add_subplot call used to create each subplot.
- gridspec_kw: Dictionary with keywords passed to the GridSpec constructor used to create the grid the subplots are placed on.

Additional keyword arguments are passed to the pyplot.figure() call.

```python
x = np.linspace(0, 2*np.pi, 400)
print(x)
y = np.sin(x**2)
print(y)
```

```python
# Create figure and axes using plt.subplots()


# Plot multiple graphs on different axes
```

```python
fig = plt.figure(figsize=(5,5))
ax = fig.subplots()

ax.plot(x,y)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1,2)
ax1.plot(x,y)
ax2.scatter(x,y)
plt.suptitle("Two Subplots Example ")
plt.show()
```

```python
fig = plt.figure(figsize=(10,10))

axes = fig.subplots(2,2)

axes[0,0].plot(x,y, label = "cos")
axes[0,0].set_title("Simple Plotr of Cos")
axes[0,0].legend()

axes[0,1].plot(np.exp(x) , label = "eponential")
axes[0,1].legend()

axes[1,0].plot(np.log(x), label = "log")
axes[1,0].legend()

axes[1,1].scatter(x,y, label = 'scatter')
axes[1,1].legend()



fig.suptitle("Different MUltiple Plots")

```

```python
x = np.linspace(0, 2*np.pi, 400)

y1 = np.sin(x**2)
y2 = np.sin(x)


fig,(ax1,ax2) = plt.subplots(2,sharex=True)
ax1.plot(x,y1, c ='red')
ax1.set_ylabel("Sin(x ** 2)")

ax2.plot(x, y2, c= 'teal')
ax2.set_ylabel("Sin x")

fig.suptitle("Subplots with Shared X axis ")
plt.show()
```

## Real-World Visualization Scenario

Example:
- Histogram 
- Scatter 
- Bar Plot 

Subplots help combine all insights in one dashboard.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample real-world dataset (student performance)
np.random.seed(42)

# Simulated data
study_hours = np.random.uniform(1, 10, 50)
marks = study_hours * 8 + np.random.normal(0, 5, 50)

subjects = ['Math', 'Science', 'English', 'Computer', 'Social']
avg_marks = [78, 85, 72, 90, 75]

# Create subplots dashboard (2 rows, 2 columns)
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# 1. Histogram - Distribution of Marks
axs[0, 0].hist(marks, bins=10)
axs[0, 0].set_title("Marks Distribution (Histogram)")
axs[0, 0].set_xlabel("Marks")
axs[0, 0].set_ylabel("Frequency")

# 2. Scatter Plot - Study Hours vs Marks
axs[0, 1].scatter(study_hours, marks)
axs[0, 1].set_title("Study Hours vs Marks (Scatter Plot)")
axs[0, 1].set_xlabel("Study Hours")
axs[0, 1].set_ylabel("Marks")

# 3. Bar Plot - Average Marks by Subject
axs[1, 0].bar(subjects, avg_marks)
axs[1, 0].set_title("Average Marks per Subject (Bar Plot)")
axs[1, 0].set_xlabel("Subjects")
axs[1, 0].set_ylabel("Average Marks")

# 4. Turn off unused subplot (optional)
axs[1, 1].axis('off')

# Adjust layout for better spacing
plt.tight_layout()

# Show dashboard
plt.show()
```
