## 1. What is Seaborn?

Seaborn is a Python data visualization library built on top of Matplotlib that is specifically designed for statistical data visualization.

It provides a high-level interface for creating informative and attractive statistical graphics with less effort and better default aesthetics.

Key Concept:
- Matplotlib → Low-level plotting library
- Seaborn → High-level statistical visualization library

Seaborn internally uses Matplotlib but simplifies complex visualization tasks and enhances readability, making it ideal for Exploratory Data Analysis (EDA).

## 2. Why Do We Use Seaborn in Data Science?

### 2.1 Better Default Visual Appearance

Matplotlib plots are functional but visually basic by default.

Seaborn automatically applies:

- Clean grid styles
- Better color palettes
- Professional visual themes
- Improved readability

This makes plots more suitable for analysis and presentations.

### 2.2 Designed for Statistical Analysis

Seaborn is specifically built to visualize statistical characteristics of data such as:
- Data distributions
- Central tendency
- Variance and spread
- Feature relationships

This makes it extremely useful during Exploratory Data Analysis (EDA), where understanding data behavior is more important than just plotting values.

### 2.3 Direct Integration with Pandas DataFrames

In real-world data analysis, datasets are handled using Pandas DataFrames.  
Seaborn works directly with DataFrame columns, allowing us to pass column names instead of manually handling arrays.

This results in:
- Cleaner syntax
- Faster analysis workflow
- Less manual data preparation for visualization

### 2.4 Less Code, More Insight

Complex visualizations that require multiple steps in Matplotlib can often be created with fewer steps in Seaborn.  
This allows Data Scientists and ML Engineers to focus more on insights rather than plot configuration.

## 3. Matplotlib vs Seaborn (Conceptual Comparison)

### 3.1 Matplotlib (Low-Level Library)

Matplotlib is the foundational plotting library in Python that provides complete control over plots and figure elements.

Characteristics:
- Highly customizable
- More manual configuration required
- Basic default styling
- Suitable for custom and production-level visualizations
- Requires more lines of code for complex plots


### 3.2 Seaborn (High-Level Statistical Library)

Seaborn is built on top of Matplotlib and is designed to simplify statistical visualization with better default aesthetics.

Characteristics:
- High-level interface
- Attractive default themes
- Built-in statistical plots
- Direct support for Pandas DataFrames
- Faster and more efficient for EDA

### 3.3 Practical Usage in Industry

In real-world Data Science projects:
- Matplotlib is used for detailed customization and final visual outputs
- Seaborn is mainly used for Exploratory Data Analysis (EDA) and statistical visualization

Most analysts start their data exploration with Seaborn because it provides faster and clearer insights into the dataset.

## Introduction to Basic Statistical Plots in Seaborn

Statistical plots help us understand how data is distributed and how values behave across a dataset.

Advanced statistical and categorical plots will be covered in the next lecture.

## Histogram – Distribution Visualization


### Concept of Histogram

A Histogram is used to visualize the distribution of a numerical variable by dividing data into bins and showing frequency counts.

It helps us analyze:
- Data distribution pattern
- Skewness (left or right)
- Concentration of values
- Presence of outliers

Histogram is one of the first plots used during Exploratory Data Analysis in real-world Data Science projects.

```python
# Create figure and axes using plt.subplots()
fig, ax = plt.subplots(figsize=(8,5))

# Select the numerical column for distribution analysis (e.g., total bill)
# Plot histogram using seaborn statistical plotting function

sns.histplot(data = df, x="total_bill", ax = ax, color="teal")

# Set title and axis labels for interpretation
ax.set_title("Histogram Plot of Total Bill Distribution")
ax.set_xlabel("Total Bills")
ax.set_ylabel("Frequency")

# Display the final plot
plt.plot()
```

## Kernel Density Plot (KDE) – Basic Introduction

### Concept of KDE Plot

KDE (Kernel Density Estimation) is a smooth curve that represents the probability density of a dataset.

Unlike histograms, which use bars, KDE provides a continuous visualization of distribution.  
This makes it easier to understand:
- Data smoothness
- Distribution peaks
- Underlying probability density
- Comparative distribution behavior

KDE plots are widely used in statistical analysis and feature distribution understanding.

```python
# Create figure and axes for KDE visualization
fig, ax = plt.subplots(figsize=(5,5))

# Select the numerical feature for density analysis
# Plot KDE curve using seaborn density plotting function
sns.kdeplot(data= df, x ="total_bill", ax = ax, fill=True)

# Add title and axis labels for better understanding
ax.set_title("Kde Plot of Total Bill")
ax.set_xlabel("total Bill")
ax.set_ylabel("Denisty")


# Display the visualization
plt.show()
```

### Scatter Plot (Relationship Visualization)


A Scatter Plot is used to visualize the relationship between two numerical variables.  
It helps us understand:
- Correlation between features
- Patterns and trends
- Clusters in data
- Outliers

This plot is very important in Data Science because it helps in identifying relationships between independent and dependent variables before model building.

```python
# Create figure and axes using plt.subplots()
fig, ax = plt.subplots()

# Select two numerical features from the dataset (e.g., total_bill and tip)
# Create a scatter plot using seaborn relationship plotting function
sns.scatterplot(data = df, x ="total_bill", y="tip", ax = ax)

# Add plot title and axis labels for interpretation
ax.set_title("Scatter Plot: Total Bill vs Tip")
ax.set_xlabel("Total Bill")
ax.set_ylabel("TIP")

# Display the final visualization
plt.plot()
```

### Count Plot (Categorical Distribution)


A Count Plot is used to show the frequency count of categorical variables.  
It helps us understand:
- Category distribution
- Class imbalance
- Most frequent categories
- Data composition

This is especially useful when working with categorical features such as gender, day, time, or category labels in real-world datasets.

```python
# Create figure and axes using plt.subplots()
fig, ax = plt.subplots()

# Select a categorical column from the dataset (e.g., day or sex)
# Create a count plot to visualize category frequencies
sns.countplot(data=df, x ="day", ax =ax )

# Set appropriate title and axis labels
ax.set_title("Count Plot of Days")
ax.set_xlabel("Day")
ax.set_ylabel("Count")

# Display the plot
plt.plot()
```

### Bar Plot (Statistical Aggregation Plot)


A Bar Plot in Seaborn is different from a simple bar chart.  
It shows the aggregated statistical value (by default mean) of a numerical variable across different categories.

It helps in:
- Comparing averages across categories
- Understanding category-wise trends
- Quick statistical comparison between groups

This is widely used in business analytics and feature analysis.

```python
# Create figure and axes using plt.subplots()
fig, ax = plt.subplots()

# Select a categorical feature for x-axis (e.g., day)
# Select a numerical feature for y-axis (e.g., total_bill)
# Create a seaborn bar plot for statistical aggregation
sns.barplot(data = df, x ="day", y = "total_bill", ax = ax)

# Add title and axis labels for clarity
ax.set_title("Average Total Bill by day")
ax.set_xlabel("Day")
ax.set_ylabel("Average Total Bill")

# Display the visualization
plt.show()
```

### Line Plot (Trend Visualization)



A Line Plot is used to visualize trends over a continuous variable or ordered data.  
It is especially useful for:
- Time-series analysis
- Trend detection
- Pattern observation across sequences
- Continuous data behavior

In statistical analysis, line plots help track how a numerical variable changes with respect to another variable.

```python
# Create figure and axes using plt.subplots()
fig, ax = plt.subplots()

# Select the feature for x-axis (continuous or ordered variable)
# Select the numerical feature for y-axis
# Create a seaborn line plot to observe trends
sns.lineplot(data = df, x ="day", y ="total_bill", ax = ax)

# Add title and axis labels for interpretation
ax.set_title("Line Plot of Total bil acrross days")
ax.set_xlabel("Day")
ax.set_ylabel(" Total Bill")

# Display the plot
plt.show()
```
