# Plotting x and y points

The `plot()` function is used to draw points (markers) in a diagram.

By default, the `plot()` function draws a line from point to point.

The function takes parameters for specifying points in the diagram.

Parameter 1 is an array containing the points on the x-axis.

Parameter 2 is an array containing the points on the y-axis.


```python
# Create sample x and y data
xpoints = np.array([1,8])
ypoints = np.array([3,10])


# Plot a basic plot
plt.plot(xpoints, ypoints)
plt.show()
```

We can use the shortcut string notation parameter to specify the marker, line style, and color in a single argument.

This parameter is called **fmt** and follows this syntax:
`marker|line|color`

## Common Customization Features

We can customize plots using:
- Title (plt.title)
- X-axis label (plt.xlabel)
- Y-axis label (plt.ylabel)
- Legends (plt.legend)
- Colors and styles


```python
months = [
    'Jan', 'Feb','Mar','Apr','May','Jun'
]
sales_on_2025 = [105000, 100000, 120000, 135000, 145000, 110000]
sales_on_2026 = [100000, 90000, 125000, 130000, 140000, 130000]
```

```python
plt.plot(months, sales_on_2025, marker='o', color ='black')
plt.plot(months, sales_on_2026, marker = 'o', color = 'teal')
plt.show()
```

```python
plt.title("Sales Comparison of  Six months on 2025 vs 2026")

plt.xlabel("Months")
plt.ylabel("Number of Sales")

plt.plot(months, sales_on_2025, marker='o', color ='black')
plt.plot(months, sales_on_2026, marker = 'o', color = 'teal')
plt.show()
```

```python
plt.title("Sales Comparison of  Six months on 2025 vs 2026")

plt.xlabel("Months")
plt.ylabel("Number of Sales")

plt.xticks(rotation = 75)
plt.yticks(rotation = 40)

plt.plot(months, sales_on_2025, marker='o', color ='black')
plt.plot(months, sales_on_2026, marker = 'o', color = 'teal')

plt.legend(['sales_on_2025', 'sales_on_2026'], loc = 'lower right')
plt.show()
```

## Adding Legends to Plots

Legends help identify multiple data series in a plot.
This is useful when comparing multiple variables.


```python
plt.title("Sales Comparison of  Six months on 2025 vs 2026")

plt.xlabel("Months")
plt.ylabel("Number of Sales")

plt.xticks(rotation = 75)
plt.yticks(rotation = 40)

plt.plot(months, sales_on_2025, marker='o', color ='black')
plt.plot(months, sales_on_2026, marker = 'o', color = 'teal')

plt.legend(['sales_on_2025', 'sales_on_2026'], loc = 'lower right')
plt.show()
```

## Line Plot (Trend Visualization)

A line plot is a type of graph that displays data points called markers connected by straight line segments. \
It is generally used to visualize the relationship between two variables; one variable on the x-axis and another on the y-axis.

A line plot is used to show trends over a sequence such as:
- Time series data
- Sales growth
- Price changes


```python
x = [1, 2,3,4,5]
y = [1,8,27,64,125]

plt.plot(x,y,'o-')

plt.title("Cube of Number ")
plt.grid()
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [10,15,7,12,8]
y2 = [8,12,6,10,15]

plt.plot(x, y1, label = 'line-1')
plt.plot(x,y2, label = 'line-2', linestyle = '--', marker ='o')
plt.xlabel('X-axis')
plt.ylabel("Y-axis")
plt.title("Multiple Line with Legends")
plt.legend()
plt.show()
```

```python
x = np.linspace(0,5,100)
y1 = x ** 2
y2 = x

plt.fill_between(x, y1, y2, color ='skyblue', alpha = 0.4, label = "Filled Area")
plt.plot(x,y1,label = 'y = x^2 ')
plt.plot(x, y2, label = 'y= x')
plt.legend()
plt.show()
```

## Bar Plot (Category Comparison)

A bar plot is a graphical representation of data where rectangular bars or columns are used to represent different categories.\
The height of each bar corresponds to the value it represents.

 
Bar plots are used to compare values across different categories.
Example:
- Sales by region
- Price by fuel type


```python
# Create category labels and corresponding values
season_df = pd.DataFrame({
    'season': ['Spring', 'Summer', 'Fall'],
    "No of Students":[12, 34, 45]
})


# Plot a basic bar chart
plt.figure(figsize = (6,6))
plt.bar(season_df['season'], season_df['No of Students'], color = ['blue', 'teal','red'], label = season_df['season'])
plt.title("Bar Graph of Season and Number of Students")
plt.xlabel("Season")
plt.ylabel("No of Students")
plt.legend()

plt.show()
```

## Histogram (Data Distribution)

A histogram is like a visual summary that shows how often different values appear in a set of data. \
Imagine you have a collection of numbers, like ages of people. \
A histogram divides these numbers into groups, called "bins," and then uses bars to represent how many numbers fall into each bin. The taller the bar, the more numbers are in that group.

Histogram shows how data is distributed across ranges.
Useful for:
- Understanding frequency distribution
- Detecting skewness
- Analyzing numerical data


```python
# Create numerical dataset
data = np.random.rand(1000)
# print(data)
plt.hist(data, bins = 30, edgecolor= 'black', density = True, alpha = 0.7, color = 'green')
plt.title("Histogram Figure")
plt.xlabel("values")
plt.ylabel("pd")
plt.show()

# Plot a histogram
```

## Scatter Plot (Relationship Between Variables)

A scatter plot is a type of graph where individual data points are plotted on a two-dimensional plane. 

Each point represents the values of two variables, with one variable on the x-axis and the other on the y-axis. 

Scatter plots are useful for visualizing the correlation between two continuous variables.



Scatter plots help visualize the relationship between two numerical variables.



```python
x = [1,3,5,7,9]
y = [10,5,15,8,12]

plt.scatter(x,y,color=['red', 'green', 'blue', 'orange', 'purple'] , sizes= [30,180,120,50,60])
plt.title('Scatter Plot')
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
x = np.array([5,6,7,8,8,8,3,4,5])
y = np.array([15,16,17,18,18,18,13,14,15])

plt.scatter(x, y)

x = np.array([15,16,17,18,18,18,13,14,15])
y = np.array([25,26,27,28,28,28,23,24,25])

plt.scatter(x,y)

plt.show()
```

## Pie Chart (Proportion of categories)

A pie chart is a circular graph that represents data in slices, where each slice corresponds to a category or portion of the whole.

The size of each slice reflects the proportion of data it represents.

It is a visual way to show how parts contribute to a whole.

A pie chart shows how a whole is divided into different parts (percentages).

Example use: monthly expense breakdown.

```python
# create labels and sized and plot pie chart

y = np.array([35,25,25,15])
plt.pie(y)
plt.show()
```

```python
y = np.array([35,25,25,15])
labels=['A','B','C','D']
explode=[0,.1,0.8,0.3]
plt.pie(y, labels = labels, startangle=90, explode = explode,colors=['red','blue','black','hotpink'])
plt.legend(title = 'ABCD')
plt.show()
```
