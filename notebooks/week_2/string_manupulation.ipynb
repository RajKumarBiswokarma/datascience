## Strings in Python

### What is a String?
A string is a sequence of **characters** used to store and manipulate text.

Key points:
- Strings are **immutable**
- Enclosed in single `' '` or double `" "` quotes
- Very common in data cleaning and preprocessing

Examples:
- Names
- Emails
- Text data


```python
# Create a string
string_1 = "Hello World"


# Print the string
print(string_1)


# Check the type of the string
print(type(string_1))

print(string_1[4])
```

## String Operations

Python provides various operations to work with strings:
- Concatenation
- Repetition
- Indexing
- Length calculation


```python
# Create two strings
string_1 = "Hello"
string_2 = "World"


# Concatenate strings
concat_string = string_1 + " "+ string_2 
print(concat_string)

# Repeat a string
print(string_1 * 5)


# Access a character using index
print(string_1[4])



# Find length of string
print(len(string_1))
```

## Common String Methods

Some frequently used string methods:
- `lower()`
- `upper()`
- `strip()`
- `replace()`
- `split()`
- `find()`

These methods are widely used in **text preprocessing**.


```python
# Create a string
string_4 = "Python Programming,  Language " # Python Programming Language
print(len(string_4))

# Convert to lowercase
print(string_4.lower())


# Convert to uppercase
print(string_4.upper())


# Remove whitespace
print(string_4.strip())
print(len(string_4.strip()))


# Replace characters
print(string_4.replace("Python", "Java"))


# Split the string
print(string_4.split())


# Find a substring
print(string_4.find("Python",0,6))
```

## String Slicing

String slicing allows extracting a portion of a string.

Syntax:

string[start : end]


Used for:
- Extracting substrings
- Text processing


```python
# Create a string
string_5 = "Python Programming Language "

string_6 = "Language"
print(string_6[1:5])



# Slice first 5 characters
print(string_5[2:5])


# Slice last 5 characters
print(string_5[-5:])
```

## Regular Expressions (Regex)

Regular Expressions are used to **search, match, and manipulate text**.

They are very powerful for:
- Pattern matching
- Data validation
- Text extraction

Python provides the `re` module for regex operations.


```python
import re 
```

## Common Regex Functions

- `re.search()` → Searches for a pattern
- `re.findall()` → Returns all matches
- `re.sub()` → Replaces matches


```python
# Define a text string
text =  " My Phone num is  9856321457 and secondary number is 9865794312"

# Search for a pattern
search_result = re.search(r"\d{10}",text)
print(search_result)


# Find all matches
all_result = re.findall(r"\d{10}", text)
print(all_result)

# Replace using regex
masked_text = re.sub(r"\d", "*", text)
print(masked_text)

```

```python
text_search = re.findall(r"[a-zA-Z][a-z ]+",text)
print(text_search)
```

# File Handling in Python  
## Read, Write, and Append Operations



## Why File Handling is Important?

File handling allows a program to:
- Store data permanently
- Read data from files
- Write results to files

In real-world applications, file handling is used for:
- Logging
- Data storage
- Reading datasets
- Report generation


## Types of File Operations

Python mainly supports the following file operations:
- Read (`r`)
- Write (`w`)
- Append (`a`)

Files are handled using the built-in `open()` function.


## Opening a File

Basic syntax:

open("filename", "mode")



Common modes:

`r` - Read - Default value. Opens a file for reading, error if the file does not exist

`a` - Append - Opens a file for appending, creates the file if it does not exist

`w` - Write - Opens a file for writing, creates the file if it does not exist

`x` - Create - Creates the specified file, returns an error if the file exists

## Writing to a File

Write mode (`w`) is used to:
- Create a new file
- Overwrite an existing file

Use:
- `write()` to write text
- `writelines()` to write multiple lines


```python
# Open file in write mode
file = open("student.txt","w")

# Write content to the file
file.write("Name: Ram \n")
file.writelines(["This is first line ", "This is second line"])

# Close the file
file.close()
```

## Reading from a File

Read mode (`r`) is used to:
- Read existing file content

Common methods:
- `read()` → reads entire file
- `readline()` → reads one line
- `readlines()` → reads all lines as a list


```python
# Open file in read mode
file = open("student.txt","r")

# Read entire content
# print(file.read())

# Read one line)
# print(file.readline())

# Read all lines
print(file.readlines())


# Close the file
file.close()
```

## Appending to a File

Append mode (`a`) is used to:
- Add new content at the end of a file
- Preserve existing data


```python
# Open file in append mode
file = open("student.txt","a")

# Append content to the file
file.write("This is new apppended line ")


# Close the file
file.close()
```
