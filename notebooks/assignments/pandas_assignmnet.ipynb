# Assignment: Pandas Data Manipulation & Business Insights

## Dataset Description
The dataset represents product inventory with the following columns:
- Category
- Price
- Rating
- Stock
- Discount

This dataset contains missing values and mixed stock information.



## 🎯 Objective

Apply Pandas data manipulation techniques to clean, transform, analyze, and extract meaningful business insights from a real-world product inventory dataset.





### 📝 Task 1: Data Loading & Initial Inspection (Real Scenario: Inventory Audit)

You are a data analyst working for an e-commerce company.

Perform the following:

- Load the dataset into a Pandas DataFrame
- Display basic information about the dataset
- Identify missing values in each column
- Summarize the data types and shape of the dataset

### 📝 Task 2: Data Cleaning (Real Scenario: Data Quality Fix)

The dataset has missing values and inconsistent records.

Clean the dataset by:

- Handling missing values in Category and Stock columns
- Identify products that are missing stock information
- Filling or removing missing Price and Discount values appropriately
- Removing duplicate records if any
- Ensuring data is ready for analysis


### 📝 Task 3: Product Stock Analysis (Business Insight)

The company wants to understand inventory status.

Using the cleaned dataset:

- Count how many products are "In Stock" and "Out of Stock"
- Calculate the average price of in-stock products
- Provide insights on inventory availability


### 📝 Task 4: Price & Discount Analysis (Revenue Strategy)

Management wants to analyze pricing strategy.

Perform the following:

- Find the most expensive and least expensive products
- Calculate average discount offered across products
- Identify products with discount greater than 30%
- Analyze relationship between price and discount


### 📝 Task 5: Category-Based Business Insights (Grouping & Aggregation)

The marketing team wants category-level insights.

Using grouping and aggregation:

- Calculate average price per category
- Calculate average rating per category
- Count number of products in each category
- Identify the best-performing category based on rating


### 📝 Task 6: Data Transformation (Filtering & Sorting)

The company wants premium product insights.

Perform the following transformations:

- Filter products with price greater than 3000
- Filter products with rating above 4.0
- Sort products by price (descending)
- Create a new column: Final Price after Discount


### 📝 Task 7: Missing Data Impact Analysis (Analytical Thinking)

Analyze how missing data affects business decisions:

- Count rows with missing Category
- Count rows with missing Stock
- Compare average price before and after cleaning
- Explain the impact of missing values on analysis accuracy


### 📝 Task 8: Customer Recommendation System (Applied Pandas Logic)

Create a recommendation logic:

- Identify top-rated products (Rating ≥ 4)
- Ensure they are "In Stock"
- Sort them by highest discount
- Display top 3 recommended products


### 📝 Task 9: Advanced Insight – Profit Opportunity Detection

Assume higher discount + high rating = high sales potential.

Analyze and find:

- Products with rating above average
- Discount above average
- In stock availability
- Highlight potential high-profit products


### 📝 Task 10: Manager Dashboard Summary (Final Business Report)

Generate a summary report including:

- Total number of products
- Average price
- Average rating
- Stock distribution
- Top category by performance

Present the summary as a structured table or DataFrame.
