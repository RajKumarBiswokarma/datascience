## What is BeautifulSoup ?

BeautifulSoup is a Python library used to parse HTML and XML documents.

It helps convert raw HTML code into a structured tree format.

This allows us to easily navigate and search HTML elements.

Key features of BeautifulSoup:

• Parse HTML documents  
• Search for specific tags  
• Extract text and attributes  
• Navigate HTML structure

## Understanding HTML Structure


Webpages are structured using HTML tags.

Common tags include:

`<html>` – Root of webpage  
`<head>` – Metadata and title  
`<body>` – Main content  
`<h1>` – Heading  
`<p>` – Paragraph  
`<a>` – Link  
`<div>` – Container  

Web scraping works by locating and extracting information from these tags.

## Parsing HTML with BeautifulSoup 

BeautifulSoup converts raw HTML text into a structured object.

This object allows us to easily search for HTML elements.

```python
# Create BeautifulSoup object
# Pass HTML content and parser type
# Print parsed HTML structure

soup = BeautifulSoup(response.text, 'html.parser')
soup
```

## Extracting the Page Title

The title tag contains the name of the webpage.

BeautifulSoup allows us to extract this information easily.

```python
# Locate the title tag
title = soup.title
# Extract title text
title_text = title.text

# Print webpage title
print("Page Title", title_text)
```

## Extracting Links from Webpages

Links on webpages are stored inside `<a>` tags.

The URL of the link is stored in the **href attribute**.

```python
# Find anchor tag
# Extract href attribute
# Print link
links = soup.find_all("a")
print(len(links))
```

```python
for link in links[:10]:
    print(link.get("href"))
```

## Extracting Multiple Elements

Webpages often contain multiple similar elements.

Example:

• multiple product titles  
• multiple news headlines  

To extract all occurrences of an element we use **find_all()**.

```python
# Find all paragraph tags
# Store results in list
# Iterate through list
# Print extracted text
book_titles = soup.find_all("h3")
book_titles
```

```python
for book in book_titles:
    print(book.text)
```

```python
book_links = []
book_title = []
```

```python
for book in book_titles:
    link = book.find("a")
    book_link = link.get("href")
    book_links.append(book_link)
    full_titles = link.get("title")
    book_title.append(full_titles)
    
```
