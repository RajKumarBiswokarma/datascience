# What is Categorical Data?

Categorical data represents labels or categories instead of numerical values.

Examples:
- Gender (Male, Female)
- Country (Nepal, India, USA)
- Species (Iris dataset)
- Day (Tips dataset)

Machine learning models cannot understand text categories directly.
They require numerical input.

# Why Encoding Categorical Data is Necessary

Encoding converts categorical values into numerical form so that:
- Machine learning models can process them
- Mathematical operations become possible
- Feature representation becomes meaningful

Without encoding, models will throw errors when encountering string values.

# Types of Categorical Data

1. Nominal Data:
- No inherent order
- Example: color, city, country

2. Ordinal Data:
- Has a meaningful order
- Example: low < medium < high

## Label Encoding

Label Encoding is a categorical encoding technique that assigns a unique integer value to each category in a feature.  
It is simple, fast, and memory-efficient but may unintentionally introduce an ordinal relationship among categories.

### Example
City Feature:
- Kathmandu → 0  
- Pokhara → 1  
- Lalitpur → 2  

### When to Use
- Categorical variable is ordinal
- There is a natural order in categories
- Tree-based models (e.g., Decision Trees, Random Forest, XGBoost)
- Categorical features with many unique values
- Situations where memory efficiency is important

### Pros
- Simple and easy to implement  
- Memory-efficient compared to One-Hot Encoding  
- Works well with tree-based algorithms  

### Cons
- Introduces implicit ordering among categories  
- Can mislead models that assume numerical relationships (e.g., Linear Regression, KNN)  
- Not suitable for nominal data when used with non-tree models

```python
# Import LabelEncoder from sklearn.preprocessing
# Initialize LabelEncoder
# Apply label encoding to categorical column
# Replace original categorical column with encoded values
from sklearn.preprocessing import LabelEncoder

data = ['Red', 'Green', 'Blue', 'Red', 'Green']

encoder = LabelEncoder()

encoded_data = encoder.fit_transform(data)
encoded_data
```

```python
data = pd.DataFrame({
    'Satisfaction': ['Low', 'High', 'Medium', 'Low', 'High'],
    'Score': [3, 8, 5, 2, 9]
})
satisfaction_order = {'Low': 0, 'Medium': 1, 'High': 2}
data['Satisfaction_Encoded'] = data['Satisfaction'].map(satisfaction_order)
print(data)
```

## One-Hot Encoding

One-Hot Encoding is a categorical encoding technique that converts each category into a separate binary column, where each column represents one category.  
It prevents the model from assuming any false order among categories.

### Example
Color Feature:
- Red → [1, 0, 0]  
- Blue → [0, 1, 0]  
- Green → [0, 0, 1]  

### When to Use
- Nominal (unordered) categorical features  
- Linear models and logistic regression  
- Number of categories is limited
- Neural networks and distance-based algorithms  
- Avoiding false ordinal relationships is important

### Pros
- Does not assume any ordinal relationship between categories  
- Widely supported in machine learning libraries  
- Easy to interpret  

### Cons
- Can lead to high dimensionality when there are many unique categories  
- Produces sparse data (many zeros)  
- Increases memory usage and computational cost for large datasets

```python
# Use pandas get_dummies or OneHotEncoder from sklearn
# Apply one-hot encoding on categorical columns
# Create new dataframe with encoded features

tips_one_hot = pd.get_dummies(tips, columns = cat_cols)
tips_one_hot.head()
```

# Dummy Variable Trap

The Dummy Variable Trap occurs when one or more dummy variables created during One-Hot Encoding are highly correlated (multicollinearity).  
This happens when all categories of a categorical feature are encoded into separate binary columns, causing redundancy in the dataset.

### Why It Happens
If a categorical variable with \(k\) categories is converted into \(k\) dummy variables, the sum of all dummy columns will always equal 1.  
This creates perfect multicollinearity because one column can be predicted using the others.

### Example
Gender Feature:
- Male → [1, 0]  
- Female → [0, 1]  

Here:

Female = 1 − Male  

This means one variable is perfectly dependent on the other, leading to multicollinearity.

### Problems Caused
- Multicollinearity in regression models  
- Unstable coefficient estimates  
- Reduced model interpretability  
- Issues in Linear Regression and Logistic Regression  

### How to Avoid the Dummy Variable Trap
- Drop one dummy column (k-1 encoding)
- Use `drop_first=True` in One-Hot Encoding
- Use regularization techniques (Ridge/Lasso)
- Prefer tree-based models (less sensitive to multicollinearity)

### When It Matters Most
- Linear Regression  
- Logistic Regression  
- Any model sensitive to multicollinearity  

Tree-based models (e.g., Decision Trees, Random Forest) are generally not affected by the dummy variable trap.

In one-hot encoding, creating all category columns can cause multicollinearity.

Solution:
Drop one dummy column (drop_first = True)

```python
# One-Hot Encoding with drop_first=True (recommended for ML models)
tips_onehot_drop = pd.get_dummies(tips, columns=cat_cols, drop_first=True)

tips_onehot_drop.head()
```

```python
print("Original Dataset Shape:", tips.shape)
print("After One-Hot Encoding Shape:", tips_one_hot.shape)
print("After One-Hot (drop_first) Shape:", tips_onehot_drop.shape)
```

## Ordinal Encoding

Ordinal Encoding is a categorical encoding technique that converts categories into integer values while preserving their natural order.  
It is suitable for features where the categories have a meaningful ranking or hierarchy.

### Example
Education Level:
- High School → 1  
- Bachelor → 2  
- Master → 3  
- PhD → 4  

### When to Use
- Ordered categorical features (e.g., ratings, education levels, satisfaction levels)
- Features with a clear ranking or hierarchy

### Pros
- Maintains the natural order of categories  
- Reduces dimensionality compared to One-Hot Encoding  
- Simple and memory efficient  

### Cons
- Not suitable for nominal (unordered) categories  
- Model may assume equal distance between categories, which may not reflect reality  
- Can introduce unintended bias if order is incorrectly assigned

```python
from sklearn.preprocessing import OrdinalEncoder
data = [['Low'], ['Medium'], ['High'], ['Medium'], ['Low']]

encoder = OrdinalEncoder(categories=[['Low', 'Medium', 'High']])
encoded_data = encoder.fit_transform(data)

print(f"Encoded Ordinal Data: {encoded_data}")
```

## Target Encoding (Mean Encoding)

Target Encoding, also known as Mean Encoding, is a categorical encoding technique where each category is replaced with the mean value of the target variable corresponding to that category.

This method incorporates target information into the encoding, making it especially useful for high-cardinality categorical features.

### Example
Category: Product ID  / Rating
- Product A → 0.75 (mean target value for A)  
- Product B → 0.42  
- Product C → 0.61  

### When to Use
- High-cardinality categorical features (e.g., ZIP codes, product IDs, user IDs)
- Supervised learning tasks where target variable is available
- Models where One-Hot Encoding would create too many columns

### Pros
- Captures the relationship between the category and the target variable  
- Reduces dimensionality compared to One-Hot Encoding  
- Effective for high-cardinality features  

### Cons
- High risk of overfitting if not handled properly  
- Requires techniques like smoothing, regularization, or cross-validation  
- Can introduce data leakage if applied before train-test split

## Frequency Encoding

Frequency Encoding is a categorical encoding technique that replaces each category with its frequency (count) or proportion of occurrence in the dataset.  
This provides a compact representation of categorical variables based on their popularity.

### Example
City Feature (Frequency):
- Kathmandu → 50  
- Pokhara → 30  
- Lalitpur → 20  

(Or as proportions: 0.50, 0.30, 0.20)

### When to Use
- High-cardinality categorical features  
- Retail, e-commerce, and clickstream data  
- Situations where category popularity carries useful information  

### Pros
- Low computational and storage requirements  
- Handles high-cardinality features efficiently  
- Simple to implement and interpret  

### Cons
- Can introduce data leakage if computed before train-test split  
- Does not capture relationship with the target variable  
- Different categories may receive similar frequencies, reducing separability

```python
import pandas as pd
data = ['Red', 'Green', 'Blue', 'Red', 'Red']
series_data = pd.Series(data)
frequency_encoding = series_data.value_counts()

encoded_data = [frequency_encoding[x] for x in data]
print("Encoded Data:", encoded_data)
```
