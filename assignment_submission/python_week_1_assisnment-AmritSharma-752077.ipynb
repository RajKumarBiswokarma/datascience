### 📝 Task 4: Student Grade Evaluation Function

Write a Python function that:

- Accepts marks of **five subjects** as input  
- Calculates the average marks  
- Returns the grade based on the average score

```python
def taking_marks():
    marks_list = []
    print("Enter marks of 5 subjects to calculate the average score:")
    for i in range(5):
        mark = int(input())

        marks_list.append(mark)

    print(f"The list of marks that you provided: {marks_list}")
    average_marks = sum(marks_list)/len(marks_list)

    print(f"The average marks is {average_marks}")
    print(grading_marks(average_marks))

def grading_marks(average_marks):
    if average_marks >= 90:
        return("Your grade is A+")

    elif average_marks >= 80:
        return("Your grade is A")

    elif average_marks >= 70: 
        return("Your grade is B+")

    elif average_marks >= 60:
        return("Your grade is B")

    elif average_marks >= 50:
        return("Your grade is C+")

    else:
        return("Your grade is Below C")

taking_marks()
```

### 📝 Task 5: Menu-Driven Calculator Using Functions

Write a Python program that:

- Uses separate functions for:
  - Addition
  - Subtraction
  - Multiplication
  - Division
- Performs the selected operation
- Repeats until the user chooses to exit


```python
def menu():
    print("""Choose from the MENU:
        1. Addition (Press A)
        2. Subtraction (Press S)
        3. Multiplication (Press M)
        4. Division (Press D)""")
    menu_select = input()
    if menu_select == 'A' or menu_select == 'a':
        addition()

    elif menu_select == 'S' or menu_select == 's':
        subtraction()

    elif menu_select == 'M' or menu_select == 'm':
        multiplication()

    elif menu_select == 'D' or menu_select == 'd':
        division()

    else:
        print("Enter a valid string from the MENU")

def addition():
    num1 = int(input("Enter first number:"))
    num2 = int(input("Enter second number:"))
    print(f"The sum is {num1 + num2}")
    check = input("Do you want to continue: y/n")
    if check == 'y' or check == 'Y':
        addition()
    
    go_menu = input("Do you want to go to the menu: y/n")
    if go_menu == 'y' or go_menu == 'Y':
        menu()

def subtraction():
    num1 = int(input("Enter first number:"))
    num2 = int(input("Enter second number:"))
    print(f"The differene is {num1 - num2}")
    check = input("Do you want to continue: y/n")
    if check == 'y' or check == 'Y':
        subtraction()
    
    go_menu = input("Do you want to go to the menu: y/n")
    if go_menu == 'y' or go_menu == 'Y':
        menu()

def multiplication():
    num1 = int(input("Enter first number:"))
    num2 = int(input("Enter second number:"))
    print(f"The product is {num1 * num2}")
    check = input("Do you want to continue: y/n")
    if check == 'y' or check == 'Y':
        multiplication()
    
    go_menu = input("Do you want to go to the menu: y/n")
    if go_menu == 'y' or go_menu == 'Y':
        menu()

def division():
    num1 = int(input("Enter first number:"))
    num2 = int(input("Enter second number:"))
    print(f"The quotient is {num1 / num2}")
    check = input("Do you want to continue: y/n")
    if check == 'y' or check == 'Y':
        division()
    
    go_menu = input("Do you want to go to the menu: y/n")
    if go_menu == 'y' or go_menu == 'Y':
        menu()
        
menu()
```

### 📝 Task 7: Safe User Input Handling

Write a Python program that:

- Takes age as input from the user  
- Handles invalid inputs such as:
  - Non-numeric values
  - Negative numbers  
- Displays appropriate error messages using exception handling


```python
#custom error handling is used here with the help of google

class InvalidAgeError(Exception):
    """Exception raised for invalid age values."""
    def __init__(self, age, message="Age must be between 0 and 120"):
        self.age = age
        super().__init__(message)
    
    def __str__(self):
        return f'{self.age} -> {self.args[0]}'
    
class NegativeValueError(Exception):
    "Exception raised for negative age value"
    def __init__(self, age, message="Age must be positive value."):
        self.age = age
        super().__init__(message)
    
    def __str__(self):
        return f'{self.age} -> {self.args[0]}'

try:
    age = int(input("Enter your age:"))
    if age < 0:
        raise NegativeValueError(age)
    if age > 120:
        raise InvalidAgeError(age)
except ValueError as ve:
    print("Error:", ve)

except NegativeValueError as nve:
    print("Error:", nve)

except InvalidAgeError as iae:
    print("Error:", iae)

else: 
    print(f"Your age is {age}")

finally:
    print("done!!!")

```

### 📝 Task 8: Banking System Error Handling

Write a Python function `withdraw()` that:

- Accepts balance and withdrawal amount as input  
- Raises an error if the withdrawal amount is greater than the balance  
- Handles invalid inputs using `try-except` blocks


```python
#suppose the user have rs 30000 already

def withdraw():
    amount = 30000
    accept_withdraw = input("'w' for withdraw and 'd' for diposit!")
    if accept_withdraw == 'w' or accept_withdraw == 'W':
        print(f"You have rs. {amount}")
        try:
            withdraw = float(input("enter the withdrawal amount"))
            if withdraw > amount:
                raise ValueError("really?? insufficient balance!!!")
            
        except ValueError as ve:
            return ve
        
        else:
            amount = amount - withdraw
            print(f"Your new balance is rs. {amount} ")
    
    elif accept_withdraw == 'd' or accept_withdraw == 'D':
        print(f"You have rs. {amount}")
        try:
            diposit = float(input("Enter how much you want to add."))
            if diposit < 0:
                raise ValueError("Negative value is not accepted.")
            
        except ValueError as ve:
            return ve
        
        else:
            amount = amount + diposit
            print(f"Your new balance is rs. {amount}")

withdraw()
```
