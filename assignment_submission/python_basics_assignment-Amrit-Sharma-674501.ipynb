### Assignment: Arithmetic Operators

Take two numbers as input from the user and:
- Calculate their sum
- Calculate their difference
- Calculate their product
- Calculate their remainder


```python
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

#Their Sum
print(num1 + num2)

# Their Difference
print(num1 - num2)

# Their Product
print(num1 * num2)

# Their remainder when divided
print(num1 % num2)
```

### Assignment: Assignment Operators

Create a variable with an initial value and use assignment operators (`+=`, `-=`, `*=`) to update the value step by step.


```python
num1 = 37

num1 += 3
print(num1)

num1 -= 2
print(num1)

num1 *= 6
print(num1)
```

### Assignment: Logical Operators

Take two boolean values and:
- Check the result using `and`
- Check the result using `or`
- Reverse one value using `not`


```python
bool1 = True
bool2 = False

print(bool1 and bool2)

print(bool1 or bool2)  

print(not bool2)

print(not bool1)

```

### Assignment: Bitwise Operators

Take two integers and perform:
- Bitwise AND
- Bitwise OR
- Bitwise XOR


```python
int1 = 20

int2 = 17

print(int1 & int2)  # Bitwise AND

print(int1 | int2)  # Bitwise OR

print(int1 ^ int2)  # Bitwise XOR

```

### Assignment: Comparison Operators

Take two numbers as input from the user and:
- Check if both numbers are equal
- Check which number is greater
- Check which number is smaller


```python
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

print(num1 == num2)  # Equality check
print(num1 > num2)   # Greater than check
print(num1 < num2)   # Less than check


if(num1 > num2):
    print(f"{num1} is greater than {num2}")
    print(f"{num2} is less than {num1}")
elif(num1 < num2):
    print(f"{num1} is less than {num2}")
    print(f"{num2} is greater than {num1}")

lambda_func = lambda num1, num2: f"{num1} is smaller" if num1 < num2  else (f"{num2} is smaller" if num1 > num2 else "both are equal")
print(lambda_func(num1, num2))
```
